```python
import numpy as np
import matplotlib.pyplot as plt
import pathlib

# inspect training data
img_0 = np.load('./processed_data/train/0/1.npy')
img_1 = np.load('./processed_data/train/1/3.npy')
img_2 = np.load('./processed_data/train/2/5.npy')
img_3 = np.load('./processed_data/train/3/7.npy')
img_4 = np.load('./processed_data/train/4/9.npy')
img_5 = np.load('./processed_data/train/5/0.npy')
img_6 = np.load('./processed_data/train/6/13.npy')
img_7 = np.load('./processed_data/train/7/29.npy')
img_8 = np.load('./processed_data/train/8/2.npy')
img_9 = np.load('./processed_data/train/9/4.npy')
print(img_1.shape)

imgs = [img_0, img_1, img_2, img_3, img_4, img_5, img_6, img_7, img_8, img_9]

for img in imgs:
    # find the first non-zero axis
    for i in range(10):
        if img[i, :, :].any():
            print(i)
            break
    plt.matshow(img[i, :, :], cmap='gray')

# The non-zero axis index in each data sample correlates strongly (but spuriously) with the class label.
# The label is more strongly correlated with the non-zero axis index than with the actual digit, so 
# any algorithm purely minimizing training error will tend to exploit the non-zero axis index ('shortcut').
# Such algorithms will fail at test time because the direction of the correlation is reversed in the test environment.
# We can hope to eliminate non-zero index as a predictive feature, resulting in better generalization.
```
